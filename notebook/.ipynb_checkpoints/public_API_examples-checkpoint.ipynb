### Load some TMT data
Extract uniprot ID with Fold Change > 2.0 on the **Abundance Ratio: (pTrc-YfjB) / (pTrc)** values

### Set ressources endpoints

```python
FLASK_ENDPOINT="http://localhost:2222/compute"
NEST_ENDPOINT="http://localhost:3000/api/pwas/ora"
```

```python
import pandas
import numpy as np

df=pandas.read_csv(tmt_data, sep="\t")
df['Abundance Ratio: (pTrc-YfjB) / (pTrc)'] = df['Abundance Ratio: (pTrc-YfjB) / (pTrc)'].str.replace(',','.')
df = df.astype({'Abundance Ratio: (pTrc-YfjB) / (pTrc)': float} )

DELTA_df = df.loc[(df['Abundance Ratio: (pTrc-YfjB) / (pTrc)'] > 1.5)]

len (DELTA_df['Accession'].tolist())
```

```python
import requests, json

data = {
    "all_accessions"           : df['Accession'].tolist(),
    "significative_accessions" : DELTA_df['Accession'].tolist(),
    "taxid"                    : "83333", 
    "method"                   : "fisher"
}

ans = requests.post(FLASK_ENDPOINT, json = data)
ans
```

```python
import requests, json

data = {
    "proteinsExp"           : df['Accession'].tolist(),
    "proteinsDelta" : DELTA_df['Accession'].tolist(),
    "taxid"                    : 83333, 
    "method"                   : "fisher"
}

ans = requests.post(NEST_ENDPOINT, json = data)
#requests.get("http://localhost:3000/api/pwas")
ans
```

```python
ora_results = json.loads(ans.text)
ora_results['scoreC']
```

```python
_ = [ d for d in ora_results['scoreC'] if d['p_val'] < 0.05 ]
```
