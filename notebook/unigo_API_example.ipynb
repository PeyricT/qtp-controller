```python
import requests
```

```python
apiEndpoint = "http://localhost:3000"
proxies = {
  'http': '',
  'https': '',
}
```

```python
uniprotIds = ["P0A6Y8", "P0A6M8"]
taxid = 83333
```

```python
dic_uniprot = {}
for prot in uniprotIds:
    uniprot_ans = requests.get(apiEndpoint + "/api/uniprot/get/" + prot, proxies=proxies).json()
    go_terms = [dic["id"] for dic in uniprot_ans[prot]["GO"]]
    dic_uniprot[prot] = set(go_terms)
```

```python
inputAPI = { "uniprotIds" : uniprotIds, "taxid" : taxid }
```

```python
ans_unigo = requests.post(apiEndpoint + "/api/unigo/trimmedVectorByProt", json = inputAPI, proxies=proxies).json()
```

```python
print(ans_unigo)
```

```python
dic_unigo = {}
for prot in ans_unigo:
    go_terms = [dic["go"] for dic in ans_unigo[prot]]
    dic_unigo[prot] = set(go_terms)
```

```python
for prot in dic_uniprot : 
    print(prot)
    intersection = dic_uniprot[prot].intersection(dic_unigo[prot])
    if len(intersection) != len(dic_uniprot[prot]):
        print(prot, "problem")
```
